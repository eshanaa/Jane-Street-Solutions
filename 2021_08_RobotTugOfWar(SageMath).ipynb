```python
y = var('y')
assume(y>-1/2)
assume(y<1/2)
assume(y,'real')

y=0

x = PolynomialRing(QQ,'x')
a = piecewise([[[0,1],1]])
b = piecewise([[[-1,0],1]])


r = piecewise([[[-y,1-y],1]])
r = r.restriction([-1/2,1/2])

print(r.integral())

for i in range(5):
    r=r.convolution(b)
    print(r.integral())
    r=r.restriction([-1/2,1/2])
    
    r=r.convolution(a)
    print(r.integral())
    r=r.restriction([-1/2,1/2])
```

```python
def ans(y):
    x = PolynomialRing(QQ,'x').gen()
    a = piecewise([[[0,1],1]],var=x)
    b = piecewise([[[-1,0],1]],var=x)
    a_tot = 0

    r = piecewise([[[-1,1],1]],var=x)
    r = r.restriction([-y,1-y])

    res = 1-r.integral()(x=1/2)
    a_tot += res
    r=r.restriction([-1/2,1/2])

    while r.integral()(x=r.end_points()[-1]-0.0000000001) > 1e-8:
        r=r.convolution(b)
        res = r.integral()(x=-1/2)
        r=r.restriction([-1/2,1/2])

        r=r.convolution(a)
        res = r.integral()(x=3/2)-r.integral()(x=1/2)
        a_tot += res
        r=r.restriction([-1/2,1/2])
        
    return a_tot
    
```

```python
y = 0.2849999
for i in range(3):
    y += 0.0000001
    print(y,ans(y))
```
