```python
import numpy as np
import time
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
from IPython.display import Markdown, display,IFrame,HTML,Image
import itertools
from matplotlib.patches import Rectangle
```

```python
url='https://www.janestreet.com/puzzles/new-york-minute-index/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
tags=soup.findAll('img')
for match in tags:
    match.decompose()

display(Markdown('### Puzzle as published \n'+url))
display(Image('https://www.janestreet.com/puzzles/new-york-minute.png',width=400))
HTML(str(soup.find_all('div', {'class' :'col-12'})[4]))
```

```python
raw_text = '\
ytimoiczryzwbus\
tegcirbpylkoorb\
retpemplompilzy\
erhighlipetzxiz\
btmloisterslicp\
isstopestreetrk\
llteczerepzucze\
almeptrzlfzrkue\
oztbropxyoolegs\
ewjakoreztowpzt\
uigrzpcmeptrzlz\
ttzlmopeyislzpc\
zerzuqsposiczμi\
tlepputcpzllohu\
srzbzyypewnseuμ'

coding = {'z':'a',
          'a':'f',
          'f':'p',
          'p':'n',
          'n':'μ',
          'μ':'m',
          'm':'c',
          'c':'d'}


mytable = raw_text.maketrans(coding)
converted_text = raw_text.translate(mytable)

text = [i for i in converted_text]
grid = np.array(text).reshape(15,15)

fig,ax = plt.subplots(1,1,figsize=(5,5))
highlights =np.zeros((15,15))

highlights[:,0]=1 
print(1,"".join(reversed(grid[highlights ==1])))
highlights[5,2:13]=2
print(2,"".join(grid[highlights ==2]))
highlights [3,2:10]=3
print(3,"".join(grid[highlights ==3]))
highlights [7,2:13]=4
print(4,"".join(grid[highlights ==4]))
highlights [2:15,14]=5
print(5,"".join(grid[highlights ==5]))
highlights [8,3:11]=6
print(6,"".join(grid[highlights ==6]),"SPELLING ERROR")
highlights [9,4:13]=7
print(7,"".join(grid[highlights ==7]))
highlights [10,2:14]=8
print(8,"".join(grid[highlights ==8]))
highlights [11,4:15]=9
print(9,"".join(grid[highlights ==9]))
highlights [12,1:14]=10
print(10,"".join(reversed(grid[highlights ==10])))
highlights [13,1:14]=11
print(11,"".join(reversed(grid[highlights ==11])))
highlights [0,9:15]=12
print(12,"".join(reversed(grid[highlights ==12])))
highlights [0,0:9]=13
print(13,"".join(reversed(grid[highlights ==13])))
highlights [1,1:15]=14
print(14,"".join(reversed(grid[highlights ==14])))
highlights [0:10,1]=15
print(15,"".join(reversed(grid[highlights ==15])))
highlights [2,0:13]=16
print(16,"".join(reversed(grid[highlights ==16])))
highlights [3,10:14]=17
print(17,"".join(grid[highlights ==17]))
highlights [4,2:11]=18
print(18,"".join(grid[highlights ==18]))
highlights [6,3:13]=19
print(19,"".join(reversed(grid[highlights ==19])))
highlights [14,7:15]=20
print(20,"".join(grid[highlights ==20]))
highlights [14,1:6]=21
print(21,"".join(reversed(grid[highlights ==21])),"SPELLING ERROR")
highlights [9,2:5]=22
print(22,"".join(grid[highlights ==22]),)
highlights [2:11,13]=23
print(23,"".join(reversed(grid[highlights ==23])))

print("\nAnswer is:","".join(grid[highlights ==0]))

ax = sns.heatmap(highlights,annot =grid,fmt='',cbar=False,cmap='tab20',linecolor='k',linewidth=2,annot_kws={"fontsize":14})
for i,j in itertools.product(range(15),repeat=2):
        if highlights[i,j]==0:
            ax.add_patch(Rectangle((j,i), 1, 1, fill=False, edgecolor='w', lw=3))
ax.axis('off')
plt.tight_layout()
plt.show()
```
